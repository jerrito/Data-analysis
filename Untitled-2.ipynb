```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot
%matplotlib inline
from scipy.stats import pearsonr
```

```python
main=pd.read_csv("Second Trimester.csv")
main.set_index(["Unnamed: 0"], inplace=True)
```

```python
main
```

```python

s=main["BPD"].str.replace("w","", regex=False)
l=s.str.rsplit("d",expand=True,n=1)
pd.DataFrame( l)

main.columns
```

```python
m=['BPD', 'HC', 'AC', 'FL', 'TCD', 'LMP']
main=main[m].replace({"w":".","d":""},regex=True)
main
```

```python
main_1=main[m].replace({". ":" "},regex=True)
main_1
```

```python
main_2=main_1['BPD'].str.split(" ", n=1, expand=True)
main_1["BPD/Weeks"]=main_2[0]
main_1["BPD/Days"]=main_2[1]
main_1
```

```python
main_1["BPD/Days"].loc[36:37,]="1"
main_1
```

```python
main_1["BPD/Days"]=pd.to_numeric( main_1["BPD/Days"])
main_1["BPD/Days"].dtype
```

```python
main_1["BPD/Weeks"]=pd.to_numeric(main_1["BPD/Weeks"])	
main_1.columns
```

```python
m3=main_1['HC'].str.split(" ", n=1, expand=True)
m4=main_1['AC'].str.split(" ", n=1, expand=True)
m5=main_1['FL'].str.split(" ", n=1, expand=True)
m6=main_1['TCD'].str.split(" ", n=1, expand=True)
m7=main_1['LMP'].str.split(" ", n=1, expand=True)
main_1['HC/Weeks']=m3[0]
main_1['AC/Weeks']=m4[0]
main_1['FL/Weeks']=m5[0]
main_1['TCD/Weeks']=m6[0]
main_1['LMP/Weeks']=m7[0]
main_1['HC/Days']=m3[1]
main_1['AC/Days']=m4[1]
main_1['FL/Days']=m5[1]
main_1['TCD/Days']=m6[1]
main_1['LMP/Days']=m7[1]
```

```python
main_1
```

```python

main_1['HC/Weeks']=pd.to_numeric( main_1['HC/Weeks'])
main_1['AC/Weeks']=pd.to_numeric(main_1['AC/Weeks'])
main_1['FL/Weeks']=pd.to_numeric(main_1['FL/Weeks'])
main_1['TCD/Weeks']=pd.to_numeric(main_1['TCD/Weeks'])
main_1['LMP/Weeks']=pd.to_numeric(main_1['LMP/Weeks'])
main_1['HC/Days']=pd.to_numeric(main_1['HC/Days'])
main_1['AC/Days']=pd.to_numeric(main_1['AC/Days'])

main_1['LMP/Days']=pd.to_numeric(main_1['LMP/Days'])

```

```python
main_1["FL/Days"].loc[36:36,]="0" 
main_1["TCD/Days"].loc[27:27,]="0" 
```

```python
main_1["FL/Days"]=pd.to_numeric(main_1["FL/Days"])
main_1["TCD/Days"]=pd.to_numeric(main_1["TCD/Days"])
```

```python
main_1
```

```python
file="New Second Trimester.xlsx"
main_1.to_excel(file)
```

```python
file_main=main_1.loc[:,['BPD/Weeks','BPD/Days','HC/Weeks','AC/Weeks','FL/Weeks','TCD/Weeks','LMP/Weeks','HC/Days','AC/Days','FL/Days','TCD/Days','LMP/Days']]
file_main
```

```python
file_main["BPD/Weeks"]=file_main["BPD/Weeks"]*7
file_main["HC/Weeks"]=file_main["HC/Weeks"]*7
file_main["AC/Weeks"]=file_main["AC/Weeks"]*7
file_main["FL/Weeks"]=file_main["FL/Weeks"]*7
file_main["TCD/Weeks"]=file_main["TCD/Weeks"]*7
file_main["LMP/Weeks"]=file_main["LMP/Weeks"]*7
```

```python
file_main
```

```python
file_main["BPD/W/d"]=file_main["BPD/Weeks"]+file_main["BPD/Days"]
file_main["HC/W/d"]=file_main["HC/Weeks"]+file_main["HC/Days"]
file_main["AC/W/d"]=file_main["AC/Weeks"]+file_main["AC/Days"]
file_main["FL/W/d"]=file_main["FL/Weeks"]+file_main["FL/Days"]
file_main["TCD/W/d"]=file_main["TCD/Weeks"]+file_main["TCD/Days"]
file_main["LMP/W/d"]=file_main["LMP/Weeks"]+file_main["LMP/Days"]
file_main
```
